```python
#print columns and see the proportion of values in sentiment
print(df.columns)
print(df['sentiment'].value_counts())
```

```python
#define keywords that are most likely to be mentionned in the "review"
positive_keywords = ["amazing", "love", "good"]
negative_keywords = ["awful", "bad", "terrible"]
```

```python
#see if keywords appear in reviews
def contains_keyword (review, keyword):
  return keyword.lower() in review.lower()
print(contains_keyword)
```

```python
#test keyword awful
test_review = df['review'].iloc[0:5]
results = test_review.apply(lambda r: contains_keyword(r, "awful"))
print(results)
```

```python
# Calculate P(Positive): the overall fraction of positive reviews
def compute_prior(df):
    return (df['sentiment'] == 'positive').mean()

prior_positive = compute_prior(df)
print(prior_positive)
```

```python
# Calculate P(keyword|Positive): keyword frequency within positive reviews only
def compute_likelihood(df, keyword):
    positive_reviews = df[df['sentiment'] == 'positive']['review']
    matches = positive_reviews.apply(lambda r: contains_keyword(r, keyword))
    return matches.mean()
```
