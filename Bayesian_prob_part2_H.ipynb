```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IMDB Dataset.csv", sep=",", engine="python", on_bad_lines="skip")
df.tail()
```

```python
df.head(5)
```

```python
#print columns and see the proportion of values in sentiment
print(df.columns)
print(df['sentiment'].value_counts())
```

```python
#define keywords that are most likely to be mentionned in the "review"
positive_keywords = ["amazing", "love", "good"]
negative_keywords = ["awful", "bad", "terrible"]
```

```python
#see if keywords appear in reviews
def contains_keyword (review, keyword):
  return keyword.lower() in review.lower()
print(contains_keyword)
```

```python
#test keyword awful
test_review = df['review'].iloc[0:5]
results = test_review.apply(lambda r: contains_keyword(r, "awful"))
print(results)
```

```python
# Calculate P(Positive): the overall fraction of positive reviews
def compute_prior(df):
    return (df['sentiment'] == 'positive').mean()

prior_positive = compute_prior(df)
print(prior_positive)
```

```python
# Calculate P(keyword|Positive): keyword frequency within positive reviews only
def compute_likelihood(df, keyword):
    positive_reviews = df[df['sentiment'] == 'positive']['review']
    matches = positive_reviews.apply(lambda r: contains_keyword(r, keyword))
    return matches.mean()
```

```python
# Calculate P(keyword): keyword frequency across all reviews regardless of sentiment
def compute_marginal(df, keyword):
    all_reviews = df['review']
    matches = all_reviews.apply(lambda r: contains_keyword(r, keyword))
    return matches.mean()
```

```python
# Apply Bayes' Theorem to get P(Positive|keyword)
def compute_posterior(prior, likelihood, marginal):
    return (likelihood * prior) / marginal
```

```python
# Loop through all keywords and compile prior, likelihood, marginal, posterior into a table
all_keywords = positive_keywords + negative_keywords
results = []

for kw in all_keywords:
    likelihood = compute_likelihood(df, kw)
    marginal = compute_marginal(df, kw)
    posterior = compute_posterior(prior_positive, likelihood, marginal)
    results.append({
        "keyword": kw,
        "P(Positive)": prior_positive,
        "P(keyword|Positive)": likelihood,
        "P(keyword)": marginal,
        "P(Positive|keyword)": posterior
    })

results_df = pd.DataFrame(results)
results_df
```
